```python
#найдем вероятность того, что письмо это спам

V = 10 #число слов в базе
words_spam = 82
words_nospam = 116


spam_pos = 12 / (12+27) #вероятность, что пришедшее письмо это спам
spam_pos
```

```python
nospam_pos = 27 / (12+27) #вероятность, что пришедшее письмо это НЕспам
nospam_pos
```

```python
#первая цифра в скобках это количество этого слова в базе
million_pos_spam = ((0 + 1)/(words_spam + V + 2))
million_pos_nospam = ((0 + 1)/(words_nospam + V + 2))
```

```python
online_pos_spam = ((9 + 1)/(words_spam + V + 2))
online_pos_nospam = ((13 + 1)/(words_nospam + V + 2))
```

```python
access_pos_spam = ((3 + 1)/(words_spam + V + 2))
access_pos_nospam = ((10 + 1)/(words_nospam + V + 2))
```

```python
cash_pos_spam = ((0 + 1)/(words_spam + V + 2))
cash_pos_nospam = ((0 + 1)/(words_nospam + V + 2))
```

```python
bill_pos_spam = ((0 + 1)/(words_spam + V + 2))
bill_pos_nospam = ((10 + 1)/(words_nospam + V + 2))
```

```python
offer_pos_spam = ((4 + 1)/(words_spam + V + 2))
offer_pos_nospam = ((5 + 1)/(words_nospam + V + 2))
```

```python
money_pos_spam = ((19 + 1)/(words_spam + V + 2))
money_pos_nospam = ((27 + 1)/(words_nospam + V + 2))
```

```python
import math
F_spam = math.log(spam_pos, math.e) + math.log(million_pos_spam, math.e) + math.log(online_pos_spam, math.e) + math.log(access_pos_spam, math.e) + math.log(cash_pos_spam, math.e) + math.log(bill_pos_spam, math.e) + math.log(offer_pos_spam, math.e) + math.log(money_pos_spam, math.e)
F_spam
```

```python
F_nospam = math.log(nospam_pos, math.e) + math.log(million_pos_nospam, math.e) + math.log(online_pos_nospam, math.e) + math.log(access_pos_nospam, math.e) + math.log(cash_pos_nospam, math.e) + math.log(bill_pos_nospam, math.e) + math.log(offer_pos_nospam, math.e) + math.log(money_pos_nospam, math.e)
F_nospam
```

```python
P_spam = 1 / (1 + math.e**(F_nospam - F_spam))
P_spam
```
